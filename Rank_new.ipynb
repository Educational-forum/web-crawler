```python
import requests
from bs4 import BeautifulSoup
from requests_html import HTMLSession

session = HTMLSession()
URL = ['http://www.findingschool.net/ranking/cate_16','http://www.findingschool.net/ranking/BrothersEducation','http://www.findingschool.net/ranking/CEDCA-Boarding-Ranking']
sel='body > div > div > div > div > div > div.col-xs-12.margin-clean-20.portlet-body > div > table > tbody > tr > td > a'





def getLinks(url):
    r = session.get(url)
    r.html.links
    r.html.absolute_links
    results = r.html.find(sel)
    mylist = []
    for result in results:
        mytext = result.text
        mylink = list(result.absolute_links)[0]
        mylist.append((mytext, mylink))
    return mylist

def getRankDetail(url):
    result={}
    res=requests.get(url)
    res.encoding= 'utf-8'
    soup=BeautifulSoup(res.text, 'html.parser')
#     print(soup)
   #result['']=soup.select('#sh_detail_info_ranking').text
   # return result
    #result=soup.find_all('div',class_="school_brief_description")
    #result['title']=soup.select('.schdetail-school-name.fswhite-font')
    result['title']=' '.join([p.text.strip().replace("\n", '').replace("    ", '') for p in soup.select('.schdetail-school-name a')])[:-12]
    result['info']=' '.join([p.text.strip().replace("\n", '').replace("    ", '') for p in soup.select('.col-xs-6')])[:-115]
    result['desc']='\n'.join([p.text.strip().replace("\n", '').replace("    ", '') for p in soup.select('.school_brief_description')])
    result['comment']='\n'.join([p.text.strip().replace("\n", '').replace("    ", '') for p in soup.select('.school_short_comments')])
    return result
    
    #result['info']='\n'.join([p.text.strip() for p in soup.select('.school_short_comments')])
    #return result

# getRankDetail('http://www.findingschool.net/Phillips-Academy-Andover')

def main():
    ranking=[]
    collegeList = getLinks(URL[])
    for college in collegeList:
        ranking.append(getRankDetail(college[1]))
    return ranking        
```

```python
def main1():
    ranking=[]
    collegeList = getLinks(URL[1])
    for college in collegeList:
        ranking.append(getRankDetail(college[1]))
    return ranking   
```

```python
columns = ['title','info','desc','comment']
df= pd.DataFrame(main(), columns=columns)
df
```
